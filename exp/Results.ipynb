```python
import pandas as pd

df = pd.read_csv("TONES_logs.csv")
df.head()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 2, figsize=(12, 4))

ax[0].plot(df["epoch"] + 1, df["train_loss"], label="Train Loss")
ax[0].plot(df["epoch"] + 1, df["valid_loss"], label="Test Loss")
ax[0].set_title("Graph of training and testing loss against epochs")
ax[0].set_ylabel("loss")
ax[0].set_xlabel("epochs")
ax[0].legend()
ax[0].grid()

ax[1].plot(df["epoch"] + 1, df["train_acc"], label="Train Acc")
ax[1].plot(df["epoch"] + 1, df["valid_acc"], label="Test Acc")
ax[1].set_title("Graph of training and testing accuracy against epochs")
ax[1].set_ylabel("accuracy")
ax[1].set_xlabel("epochs")
ax[1].legend()
ax[1].grid()

plt.show()
```
