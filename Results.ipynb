```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_training_logs(df):
    fig, ax = plt.subplots(1, 2, figsize=(12, 4))

    ax[0].plot(df["epoch"] + 1, df["train_loss"], label="Train Loss")
    ax[0].plot(df["epoch"] + 1, df["valid_loss"], label="Test Loss")
    ax[0].set_title("Graph of training and testing loss against epochs")
    ax[0].set_ylabel("loss")
    ax[0].set_xlabel("epochs")
    ax[0].legend()
    ax[0].grid()

    ax[1].plot(df["epoch"] + 1, df["tone_train_acc"], label="Train Acc (Tone)")
    ax[1].plot(df["epoch"] + 1, df["tone_valid_acc"], label="Test Acc (Tone)")
    ax[1].plot(df["epoch"] + 1, df["pinyin_train_acc"], label="Train Acc (Pinyin)")
    ax[1].plot(df["epoch"] + 1, df["pinyin_valid_acc"], label="Test Acc (Pinyin)")
    ax[1].set_title("Graph of training and testing accuracy against epochs")
    ax[1].set_ylabel("accuracy")
    ax[1].set_xlabel("epochs")
    ax[1].legend()
    ax[1].grid()

    plt.show()
```

```python

df = pd.read_csv("exp/logs/melspectrogram/MULTITASK_logs.csv")
df.head()
```

```python
plot_training_logs(df)
```

```python
df = pd.read_csv("exp/logs/mfcc/MULTITASK_logs.csv")
df.head()
```

```python
plot_training_logs(df)
```
